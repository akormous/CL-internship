```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Closing Prices of 30 Stocks ( 10 from each cap )

Small Cap - ASHOKA, BAJAJELEC, FORTIS, IDFC, JKLAKSHMI, PVR, RAYMOND, , RCOM, VENKEYS, WELSPUNIND

Mid Cap - APOLLOTYRE, BERGEPAINT, CASTROLIND, IDBI, MUTHOOTFIN, PNB, RBLBANK, SUNTV, TATAPOWER, VOLTAS

Large Cap - ADANIPORTS, CIPLA, EICHERMOT, HDFC, HEROMOTOCO, INFY, ITC, LT, TCS, TITAN

```python
dx = pd.read_csv('Small_Cap/ASHOKA.csv')
dx.Date
```

```python
# 5.1 Importing Closing Prices of 30 stocks in one dataframe
# Reading stocks of Small Cap

all_list = {}

# Importing Small Cap Stocks
stocks_list = ['ASHOKA', 'BAJAJELEC', 'FORTIS', 'IDFC', 'JKLAKSHMI', 'PVR', 'RAYMOND', 'RCOM', 'VENKEYS', 'WELSPUNIND']
for filename in stocks_list:
    filename = filename + '.csv'
    
    # Importing the csv
    temp = pd.read_csv('Small_Cap/'+filename)
    
    # Dropping all rows where series not equal to 'EQ'
    temp = temp[temp['Series'] == 'EQ']
    
    x = {filename[:-4]:temp['Close Price'].tolist()}
    all_list.update(x)
# End Small Cap Loop

# Importing Mid Cap Stocks
stocks_list = ['APOLLOTYRE', 'BERGEPAINT', 'CASTROLIND', 'IDBI', 'MUTHOOTFIN', 'PNB', 'RBLBANK', 'SUNTV', 'TATAPOWER', 'VOLTAS']
for filename in stocks_list:
    filename = filename + '.csv'
    
    # Importing the csv
    temp = pd.read_csv('Mid_Cap/'+filename)
    
    # Dropping all rows where series not equal to 'EQ'
    temp = temp[temp['Series'] == 'EQ']
    
    x = {filename[:-4]:temp['Close Price'].tolist()}
    all_list.update(x)
# End Mid Cap Loop



# Importing Large Cap Stocks
stocks_list = ['ADANIPORTS', 'CIPLA', 'EICHERMOT', 'HDFC', 'HEROMOTOCO', 'INFY', 'ITC', 'LT', 'TCS', 'TITAN']
for filename in stocks_list:
    filename = filename + '.csv'
    
    # Importing the csv
    temp = pd.read_csv('Large_Cap/'+filename)
    
    # Dropping all rows where series not equal to 'EQ'
    temp = temp[temp['Series'] == 'EQ']
    
    x = {filename[:-4]:temp['Close Price'].tolist()}
    all_list.update(x)
# End Large Cap Loop

df = pd.DataFrame(all_list)
```

```python
df.head()
```

```python
# 6.2 Calculating average annual returns
avg_returns = df.pct_change().mean() * 252
```

```python
# Calculating average annual volatility
an_volts = df.pct_change().std() * np.sqrt(252)
```

```python
x = []
for a in range(30):
    x.append(a)
dd = pd.DataFrame(data = [avg_returns,an_volts])
dd = dd.T
dd = dd.rename(columns = {0:'Mean_Returns',1:'Mean_Volatility'})
dd = dd.reset_index()
dd
```

```python
# Fetching data in one variable
X = dd.iloc[:,[1,2]].values
```

```python
# 6.3 Applying K-Means Clustering 
# Finding the optimal number of clusters using the Elbow Method
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    
    wcss.append(kmeans.inertia_)
    
# Plotting the Elbow Curve
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS Values')
plt.show()
```

```python
# 6.4
# Applying K_Means with optimal number of Clusters

kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10)

y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```

```python
for i in X[y_kmeans == 0,0]:
    for j in range(len(dd['index'])):
        if dd.at[j,'Mean_Returns'] == i:
            print(dd.at[j,'index'])
```

```python
# Visualizing the results
plt.rcParams["figure.figsize"]=15,8

plt.scatter(X[y_kmeans == 0,0], X[y_kmeans == 0,1], s = 10, c = 'red', label = 'Cluster 1')
for k in range(len(X[y_kmeans == 0,0]) ):
    for l in range(len(dd['index'])):
        if dd.at[l,'Mean_Returns'] == X[y_kmeans == 0,0][k]:
            plt.text(X[y_kmeans == 0,0][k],X[y_kmeans == 0,1][k],dd.at[l,'index'],fontsize = 10)

plt.scatter(X[y_kmeans == 1,0], X[y_kmeans == 1,1], s = 10, c = 'blue', label = 'Cluster 2')
for k in range(len(X[y_kmeans == 1,0]) ):
    for l in range(len(dd['index'])):
        if dd.at[l,'Mean_Returns'] == X[y_kmeans == 1,0][k]:
            plt.text(X[y_kmeans == 1,0][k],X[y_kmeans == 1,1][k],dd.at[l,'index'],fontsize = 10)
            
plt.scatter(X[y_kmeans == 2,0], X[y_kmeans == 2,1], s = 10, c = 'green', label = 'Cluster 3')
for k in range(len(X[y_kmeans == 2,0]) ):
    for l in range(len(dd['index'])):
        if dd.at[l,'Mean_Returns'] == X[y_kmeans == 2,0][k]:
            plt.text(X[y_kmeans == 2,0][k],X[y_kmeans == 2,1][k],dd.at[l,'index'],fontsize = 10)

plt.scatter(X[y_kmeans == 3,0], X[y_kmeans == 3,1], s = 10, c = 'black', label = 'Cluster 4')
for k in range(len(X[y_kmeans == 3,0]) ):
    for l in range(len(dd['index'])):
        if dd.at[l,'Mean_Returns'] == X[y_kmeans == 3,0][k]:
            plt.text(X[y_kmeans == 3,0][k],X[y_kmeans == 3,1][k],dd.at[l,'index'],fontsize = 10)
            
plt.scatter(X[y_kmeans == 4,0], X[y_kmeans == 4,1], s = 10, c = 'purple', label = 'Cluster 5')
for k in range(len(X[y_kmeans == 4,0]) ):
    for l in range(len(dd['index'])):
        if dd.at[l,'Mean_Returns'] == X[y_kmeans == 4,0][k]:
            plt.text(X[y_kmeans == 4,0][k],X[y_kmeans == 4,1][k],dd.at[l,'index'],fontsize = 10)

#plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 30, c = 'cyan', label = 'Centroids')

plt.title('Cluster of Stocks')
plt.legend()
plt.show()
```
